## Initialising and invoking a model

````python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="claude-sonnet-4-6")
````

Once installed, you can import modules into your Python code using the `import` statement. There are a few common ways:

1.  **Import the entire module:**
    ```python
    import module_name
    # Then access its contents using module_name.function()
    ```

2.  **Import with an alias:**
    ```python
    import module_name as mn
    # Then access its contents using mn.function()
    ```

3.  **Import specific functions or classes:**
    ```python
    from module_name import function_name, ClassName
    # Then use function_name() or ClassName() directly
    ```

## Customising your Model

````python
model = init_chat_model(
    model="gpt-5-nano",
    # Kwargs passed to the model:
    temperature=1.0
)

response = model.invoke("What's the capital of the Moon?")
print(response.content)
````

````python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")

response = model.invoke("What's the capital of the Moon?")
print(response.content)
````

## Initialising and invoking an agent

````python
from langchain.agents import create_agent

agent = create_agent(model=model)
````

````python
agent = create_agent(model="claude-sonnet-4-5")
````

````python
agent = create_agent("gpt-5-nano")
````

````python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
````

````python
from pprint import pprint

pprint(response)
````

````python
print(response['messages'][-1].content)
````

````python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),
    AIMessage(content="The capital of the Moon is Luna City."),
    HumanMessage(content="Interesting, tell me more about Luna City")]}
)

pprint(response)
````

## Streaming Output

````python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token

    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
````
